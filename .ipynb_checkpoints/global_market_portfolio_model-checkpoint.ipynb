# Introduction

This is a simple attempt to construct a Global Market Portfolio (GMP) using only Vanguard ETFs. I will consider the investible universe to consit of stocks and bonds. Real estate is not included because the value of publicly traded real estate is very small relative to both the total value of global real estate and as a share of publicly traded assets.

### Vanguard Total World Stock ETF (VT)
The Vanguard Total World Stock ETF (VT) attempts to reflect the performance of all publicly traded stocks globally. It tracks the FTSE Global All Cap Index. As of April 2019, this index had a market cap of:

<center> $52,885,840M </center>


### Vanguard Total World Bond ETF (BNDW)
The Vanguard Total World Bond ETF (BNDW) tracks the performance of the Bloomberg Barclays Global Aggregate Float Adjusted Composite Index. Unfortunatley, the market cap of this index is not publicly available. However, in their November 2017 paper *Historical Returns of the Market Portfolio*, Doeswijk, Lam, and Swinkels estimate that global bonds had a market cap 15% higher than that of stocks at the end of 2015 (see Table 6, paper is in the repository). Using that estimate and our estimate of equity market cap above, we can estimate the global bond market cap.

VT was 76% of it's 4/26/2019 value on 12/31/2015:

$$57.62 \div 75.54 = .763$$

This implies that the market cap of VT at the end of 2015 was $40.4T:

$$.763 \times 52,885,840 = 40,351,896$$

And that the market cap of global bonds was $46.4T at the end of 2015:

$$40,351,896 \times 1.15 = 46,404,680$$

BNDW did not exist at the end of 2015. However, its two components, BNDX and BND, did. BNDW is composed of 52% BNDX and 48% BND. Using prices of BND and BNDX in December 2015 and April 2019 and those weights, we see that BNDW would have grown by 3% over that period:

$$(55.78 \div 52.88) \times .52 + (81.01/80.76) \times .48 = 1.03$$

Which implies an April 2019 BNDW market cap of $47.8T:

$$1.03 \times 46,404,680 = 47,796,820$$

### The Global Market Portfolio

Together, these imply the following weights of the GMP at the end of April 2019:


|  Fund  |       Asset Class       |    Market Cap   | GMP Weight |
|:------:|:-----------------------:|:---------------:|:----------:|
| VT     | Stocks                  | 52,885,840      | 53%        |
| BNDW   | Bonds                   | 47,796,820      | 47%        |

# Portfolio Construction

As of 1/11/2020, VT's beta to the S&P 500 was 0.94. https://sg.finance.yahoo.com/quote/VT?p=VT

Damodaran 2019 pg 117-119 explains that the current implied equity premium is the most predictive, and assumes market efficiency. Using "Trailing 12 month cash yield", on Jan 1st the S&P 500 ERP was 5.43%
http://pages.stern.nyu.edu/~adamodar/

As of 1/11/2020, the 10-year treasury bond rate (risk free rate) was 1.822
https://www.wsj.com/market-data/bonds

### Calculate GMP Expected Return

```python
#Expected VT return = S&P500 ERP * Beta + Rf
Rf = 0.01822
Re_vt = 0.0543 * 0.94 + Rf
print('VT Expected Return:', round(Re_vt,3))
```

BNDW's YTM as of 12/1/2019 is 1.4%, and has an average duration of 7.3 years 
https://investor.vanguard.com/etf/profile/portfolio/bndw

Vanguard Intermediate-Term Treasury ETF (VGIT) has a YTM of 1.7% and an average duration of 5.2 years
https://investor.vanguard.com/etf/profile/portfolio/vgit

Vanguard Long-Term Treasury ETF (VGLT) has a YTM of 2.2% and an average duration of 17.9 years

```python
# BNDW Risk Premium:

# same duration risk-free rate:

# [BNDW Dur] = x * [VGIT Dur] + (1 - x) * [VGLT Dur]
# = x[VGIT] + [VGLT] - x[VGLT]
# [BNDW] - VGLT = x * (VGIT - VGLT)
# x = ([BNDW] - [VGLT]) / ([VGIT - VGLT])

dur_bndw = 7.3
dur_vgit = 5.2
dur_vglt = 17.9
x = (dur_bndw - dur_vglt) / (dur_vgit - dur_vglt)
#x * dur_vgit + (1-x) * dur_vglt #should equal dur_bndw

ytm_bndw = .014
ytm_vgit = .017
ytm_vglt = .022

Rf_bond = x * ytm_vgit + (1-x) * ytm_vglt

# BNDW risk premium and expected return:
bndw_risk_prem = ytm_bndw - Rf_bond
Re_bndw = bndw_risk_prem + Rf
print('BNDW Expected Return:', round(Re_bndw,3))
```

```python
# GMP expected Return
Re_gmp = vt_wt_cur * Re_vt + (1 - vt_wt_cur) * Re_bndw
print('Current GMP Expected Return:', round(Re_gmp,3))
```

### Build Portfolios

```python
from scipy.stats import norm

#Re_gmp = .09 #for testing

inf = .0167 #expected inflation rate; use 5-year TIPS spread; as of 1/20/2020
Re_cash = 0.017 #current Marcus savings account rate; as of 12/5
var_GMP = 0.014457
var_cash = 0.000554
cov = 0.000645
years = 2.5

c = pd.DataFrame({'GMP weight': np.linspace(0,1,21)})
c['expected annual return'] = round(c['GMP weight'] * (Re_gmp - inf) + (1-c['GMP weight']) * (Re_cash - inf),4)
c['portfolio standard dev'] = (c['GMP weight']**2 * var_GMP +
                           (1-c['GMP weight'])**2 * var_cash +
                           2*c['GMP weight']*(1-c['GMP weight'])*cov)**0.5
c['odds of losing value'] = (norm.cdf(0,loc=c['expected annual return']
                                           ,scale=c['portfolio standard dev'])**years)
c['25th pct total return'] = ((1+norm.ppf(0.25,loc=c['expected annual return'],scale=c['portfolio standard dev']))**years)-1
c['50th pct total return'] = ((1+norm.ppf(0.50,loc=c['expected annual return'],scale=c['portfolio standard dev']))**years)-1
c['75th pct total return'] = ((1+norm.ppf(0.75,loc=c['expected annual return'],scale=c['portfolio standard dev']))**years)-1
c[['GMP weight', 'expected annual return', 'odds of losing value', '25th pct total return',
  '50th pct total return','75th pct total return']]
```

```python
money = 60000
GMP_wt = 0.2

print('Cash:', round((1 - GMP_wt) * money,2))
print('VT Investment:', round(GMP_wt * money * vt_wt_cur,2))
print('BNDW Investment:', round(GMP_wt * money * (1 - vt_wt_cur),2))
```
