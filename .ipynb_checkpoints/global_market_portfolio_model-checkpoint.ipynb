# Introduction

Investment theory argues that, assuming market efficiency, the optimal portfolio is a market-cap weighted portfolio of all investible assets. This is a simple attempt to construct such a "Global Market Portfolio" (GMP) using only Vanguard ETFs.

I will consider the investible universe to consit of stocks and bonds. Real estate is not included because the value of publicly traded real estate is very small relative to both the total value of global real estate and as a share of publicly traded assets.

To estimate market cap weights, I will use the market capitalization of the index that each Vanguard ETF tracks. There are several challenges in calculating these weights, however, which are outlined below.

### Vanguard Total World Stock ETF (VT)
The Vanguard Total World Stock ETF (VT) reflects the performance of all publicly traded stocks globally. It tracks the FTSE Global All Cap Index. As of December 2019, this index had a market cap of:

$$$57,105,259M$$


### Vanguard Total World Bond ETF (BNDW)
The Vanguard Total World Bond ETF (BNDW) reflects the performance of all publicly traded bonds globally. It tracks the performance of the Bloomberg Barclays Global Aggregate Float Adjusted Composite Index. Unfortunatley, the market cap of this index is not publicly available. However, in their November 2017 paper *Historical Returns of the Market Portfolio*, Doeswijk, Lam, and Swinkels estimate that global bonds had a market cap 15% higher than that of stocks at the end of 2015 (see Table 6). We can combine this with the equity market cap at the end of 2015 to get a 2015 global bond market cap, and then use BNDW's growth since 2015 to update it.

#### Equity market cap at end of 2015
Old market cap values for the equity index were not available at the time of writing. However, using the growth values in the *FTSE Global All Cap Index Fact Sheet* for December 2019, we can calculate the market cap at the end of 2015:

$$57,105,259 \times \frac{1}{1 + 0.271} \times \frac{1}{1 - 0.096} \times \frac{1}{1 + 0.244} \times \frac{1}{1 + 0.09} = 36,653,483$$

#### Bond market cap at end of 2015
Adding 15% gives us the bond market cap:

$$36,653,483 \times 1.15 = 42,151,505$$

#### Bond market cap at end of 2019
BNDW did not exist at the end of 2015. However, its two components, BNDX and BND, did. BNDW is composed of 52% BNDX and 48% BND. Using prices of BND and BNDX in December 2015 and April 2019 and those weights, we see that BNDW would have grown by 6% over that period:

$$\frac {56.89}{52.88} \times .52 + \frac {84.27}{80.76} \times .48 = 1.06$$

Which implies a December 2019 bond market cap of $44.7T:

$$1.06 \times 42,151,505 = 44,680,595$$

### The Global Market Portfolio

Together, these imply the following weights of the GMP at the end of December 2019:


|  Fund  |       Asset Class       |    Market Cap   | GMP Weight |
|:------:|:-----------------------:|:---------------:|:----------:|
| VT     | Stocks                  | 57,105,259      | 56%        |
| BNDW   | Bonds                   | 44,680,595      | 44%        |

# Portfolio Construction

As of 1/11/2020, VT's beta to the S&P 500 was 0.94. https://sg.finance.yahoo.com/quote/VT?p=VT

Damodaran 2019 pg 117-119 explains that the current implied equity premium is the most predictive, and assumes market efficiency. Using "Trailing 12 month cash yield", on Jan 1st the S&P 500 ERP was 5.43%
http://pages.stern.nyu.edu/~adamodar/

As of 1/11/2020, the 10-year treasury bond rate (risk free rate) was 1.822
https://www.wsj.com/market-data/bonds

### Calculate GMP Expected Return

```python
#Expected VT return = S&P500 ERP * Beta + Rf
Rf = 0.01822
Re_vt = 0.0543 * 0.94 + Rf
print('VT Expected Return:', round(Re_vt,3))
```

BNDW's YTM as of 12/1/2019 is 1.4%, and has an average duration of 7.3 years 
https://investor.vanguard.com/etf/profile/portfolio/bndw

Vanguard Intermediate-Term Treasury ETF (VGIT) has a YTM of 1.7% and an average duration of 5.2 years
https://investor.vanguard.com/etf/profile/portfolio/vgit

Vanguard Long-Term Treasury ETF (VGLT) has a YTM of 2.2% and an average duration of 17.9 years

```python
# BNDW Risk Premium:

# same duration risk-free rate:

# [BNDW Dur] = x * [VGIT Dur] + (1 - x) * [VGLT Dur]
# = x[VGIT] + [VGLT] - x[VGLT]
# [BNDW] - VGLT = x * (VGIT - VGLT)
# x = ([BNDW] - [VGLT]) / ([VGIT - VGLT])

dur_bndw = 7.3
dur_vgit = 5.2
dur_vglt = 17.9
x = (dur_bndw - dur_vglt) / (dur_vgit - dur_vglt)
#x * dur_vgit + (1-x) * dur_vglt #should equal dur_bndw

ytm_bndw = .014
ytm_vgit = .017
ytm_vglt = .022

Rf_bond = x * ytm_vgit + (1-x) * ytm_vglt

# BNDW risk premium and expected return:
bndw_risk_prem = ytm_bndw - Rf_bond
Re_bndw = bndw_risk_prem + Rf
print('BNDW Expected Return:', round(Re_bndw,3))
```

```python
# GMP expected Return
Re_gmp = vt_wt_cur * Re_vt + (1 - vt_wt_cur) * Re_bndw
print('Current GMP Expected Return:', round(Re_gmp,3))
```

### Build Portfolios

```python
from scipy.stats import norm

#Re_gmp = .09 #for testing

inf = .0167 #expected inflation rate; use 5-year TIPS spread; as of 1/20/2020
Re_cash = 0.017 #current Marcus savings account rate; as of 12/5
var_GMP = 0.014457
var_cash = 0.000554
cov = 0.000645
years = 2.5

c = pd.DataFrame({'GMP weight': np.linspace(0,1,21)})
c['expected annual return'] = round(c['GMP weight'] * (Re_gmp - inf) + (1-c['GMP weight']) * (Re_cash - inf),4)
c['portfolio standard dev'] = (c['GMP weight']**2 * var_GMP +
                           (1-c['GMP weight'])**2 * var_cash +
                           2*c['GMP weight']*(1-c['GMP weight'])*cov)**0.5
c['odds of losing value'] = (norm.cdf(0,loc=c['expected annual return']
                                           ,scale=c['portfolio standard dev'])**years)
c['25th pct total return'] = ((1+norm.ppf(0.25,loc=c['expected annual return'],scale=c['portfolio standard dev']))**years)-1
c['50th pct total return'] = ((1+norm.ppf(0.50,loc=c['expected annual return'],scale=c['portfolio standard dev']))**years)-1
c['75th pct total return'] = ((1+norm.ppf(0.75,loc=c['expected annual return'],scale=c['portfolio standard dev']))**years)-1
c[['GMP weight', 'expected annual return', 'odds of losing value', '25th pct total return',
  '50th pct total return','75th pct total return']]
```

```python
money = 60000
GMP_wt = 0.2

print('Cash:', round((1 - GMP_wt) * money,2))
print('VT Investment:', round(GMP_wt * money * vt_wt_cur,2))
print('BNDW Investment:', round(GMP_wt * money * (1 - vt_wt_cur),2))
```
